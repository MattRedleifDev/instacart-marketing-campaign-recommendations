### Assign price ranges for products

```python
# FUNCTION: Test for and assign price category to each product
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else:
        return 'Not enough data'
```

```python
# Assign price range values in new column 'price_range'
df_orders_products_combined_final['price_range'] = df_orders_products_combined_final.apply(price_label, axis = 1)
```

### Step 4 - Add new field: Busiest periods of day for sales as 'busiest_hours'

```python
# Show frequency of values in order_hour_of_day column
df_orders_products_combined_final['order_hour_of_day'].value_counts(dropna = False)
```

### Create busy_hour designation

```python
# Create a new list to contain Day type values
busy_hour = []

# Define hours according to how busy
most_busy_hour = [10, 11, 14, 15]
least_busy_hour = [5, 2, 4, 3]

# Populate result list
for value in df_orders_products_combined_final["order_hour_of_day"]:
    if value in most_busy_hour:
        busy_hour.append("Busiest hours")
    elif value in least_busy_hour:
        busy_hour.append("Least busy hours")
    else:
        busy_hour.append("Regularly busy hours")
```

```python
# Create new field busiest_hours and write result list
df_orders_products_combined_final['busiest_hours'] = busy_hour
```

```python
# Show df_orders_products_combined_final to confirm good write
df_orders_products_combined_final.head()
```

### Create two data sets with 70/30 split

```python
# Create set of random numbers
# Seed with constant
np.random.seed(4)
```

```python
# Create new list and store boolean for 70/30 split
dev = np.random.rand(len(df_ords_prods)) <= 0.7
```

```python
# Confirm dev was populated
dev
```

### Create two dataframes: big - 70%  small - 30%

```python
# Create a dataframe for 70% of the values
big = df_ords_prods[dev]
```

```python
# Create a dataframe for 30% of the values
# Note: ~ == NOT here
small = df_ords_prods[~dev]
```
